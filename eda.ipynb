```python
import pandas as pd
import numpy as np
from src.dataset import load_n_items
```

```python
tgt_markets = ["t1", "t2"]
src_markets = ["s1", "s2", "s3"]
```

```python
tgt_market = "t1"
id_bank = load_n_items(f'DATA/{tgt_market}/train_5core.tsv', 'DATA/t1/valid_run.tsv')
tgt_items = id_bank.item_id_index.keys()
```

```python
src_market = "s1"
train_src = pd.read_csv(f"DATA/{src_market}/train_5core.tsv", sep="\t")
train_src
```

```python
include_user = set(train_src[train_src["itemId"].isin(tgt_items)]["userId"])
train_src_tgt = train_src[train_src["userId"].isin(include_user)]
exclude_item = (train_src_tgt["itemId"].value_counts() <= 2).reset_index()
exclude_item = set(exclude_item[exclude_item["itemId"]]["index"].values) - tgt_items
```

```python
train_src_tgt = train_src_tgt[~train_src_tgt["itemId"].isin(exclude_item)]
train_src_tgt
```

```python
train_src_tgt.to_csv(f"DATA/{src_market}/train_5core_{tgt_market}.tsv", sep="\t", index=False)
```

```python
for tgt_market in tgt_markets:
    for src_market in src_markets:
        id_bank = load_n_items(f'DATA/{tgt_market}/train_5core.tsv', 'DATA/t1/valid_run.tsv')
        tgt_items = id_bank.item_id_index.keys()
        train_src = pd.read_csv(f"DATA/{src_market}/train_5core.tsv", sep="\t")

        include_user = set(train_src[train_src["itemId"].isin(tgt_items)]["userId"])
        train_src_tgt = train_src[train_src["userId"].isin(include_user)]
        exclude_item = (train_src_tgt["itemId"].value_counts() <= 3).reset_index()
        exclude_item = set(exclude_item[exclude_item["itemId"]]["index"].values) - tgt_items

        train_src_tgt = train_src_tgt[~train_src_tgt["itemId"].isin(exclude_item)]
        train_src_tgt.to_csv(f"DATA/{src_market}/train_5core_{tgt_market}.tsv", sep="\t", index=False)
```
